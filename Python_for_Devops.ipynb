1. Mask a phone number
Input: "0551234567" → Output: "******4567" (keep last 4 digits).

```python
ph_num = input("what is your phone number?")
y= len(ph_num)
print(f"your masked phone number is: {"*"*(y-4) + ph_num[-4:]}")
```

2. Count digit types
Given a string, count how many are digits, how many are letters, how many are other characters.

```python
count_inp= input("enter anything")
y=0
x=0
z=0
for i in count_inp:
    if i.isdigit():
      y+=1
    elif i.isalpha():
      x+=1
    else:
      z+=1
print(f"number of digits in ur input = {y} and number of letters = {x} and number of other char = {z}")
```
