1. Mask a phone number
Input: "0551234567" → Output: "******4567" (keep last 4 digits).

```python
ph_num = input("what is your phone number?")
y= len(ph_num)
print(f"your masked phone number is: {"*"*(y-4) + ph_num[-4:]}")
```

2. Count digit types
Given a string, count how many are digits, how many are letters, how many are other characters.

```python
count_inp= input("enter anything")
y=0
x=0
z=0
for i in count_inp:
    if i.isdigit():
      y+=1
    elif i.isalpha():
      x+=1
    else:
      z+=1
print(f"number of digits in ur input = {y} and number of letters = {x} and number of other char = {z}")
```

Remove duplicates while preserving order
"0551122333" → "05123" (first occurrence only).

```python
num_dup= input("enter dup number")
seen = set()   #set is a list that doesn't allow duplicates inside

for i in num_dup:
  if i not in seen:
    seen.add(i)
    print(i, end="")
```

Find repeated substrings (length 3)
For "05512345123", find all length-3 substrings that appear more than once.

```python
num_dup= input("enter dup number")
seen = set()
i=0
y=0
count =0
for i in num_dup:
  for y in num_dup:
    if i ==y:
      count+=1
  if count>1:
    if i not in seen:
      seen.add(i)
      print(i, count)
  count=0
```

All unique substrings
Generate all unique contiguous substrings of a string and return the count.

```python
num_dup= input("enter dup number")
seen = set()
i=0
y=0
count =0
for i in num_dup:
  for y in num_dup:
    if i ==y:
      count+=1
  if count==1:
    if i not in seen:
      seen.add(i)
  count=0
print(f"{seen}, count= {len(seen)}")
```
